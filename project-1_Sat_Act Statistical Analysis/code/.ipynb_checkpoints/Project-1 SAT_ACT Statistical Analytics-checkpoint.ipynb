##### 12. Write a function using only list comprehensions, no loops, to compute standard deviation. Using this function, calculate the standard deviation of each numeric column in both data sets. Add these to a list called `sd`.

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

```python
# apply the List comprehension
sd=[np.std(sat_act[column]) for column in sat_act.columns[1:]]
sd
```

```python
def standard_deviation(lst):
    """ tocalculate the standard deviation of the list"""
    mean_value =np.mean(lst)
    inside_root = sum([(i-mean_value)**2 for i in lst])/len(lst)
    return np.sqrt(inside_root)
    
```

```python
sd = sat_act.iloc[:,1:].apply(standard_deviation)
sd
```

### [0:50]Inorder to pull the data excluding the standard deviation row

```python
import scipy.stats as stats

SAT_part=sat_act['SAT Participation']
ACT_part=sat_act['ACT Participation']

stats.ttest_ind(SAT_part, ACT_part).pvalue<.05
```

```python
# we will use the  two sample t test
stats.ttest_ind(SAT_part,ACT_part)
```

Conclusion:  Because the  p value is less than the alpha value, we fail to reject the null hypothesis and we can conclude that the two population means are not equal is True and we accepeted the Alternative hypothesis.

##### 27. Generate and interpret 95% confidence intervals for SAT and ACT participation rates.

```python
# Sat
mean=np.mean(SAT_part)
sigma=np.std(SAT_part, ddof=1)
n=len(SAT_part)

stats.norm.interval(alpha=0.95, 
                           loc=mean, 
                           scale=sigma / np.sqrt(n))
```

conclusion: the True mean of act test participation falls between 0.56 and 0.74

```python
# act
mean=np.mean(ACT_part)
sigma=np.std(ACT_part, ddof=1)
n=len(ACT_part)

stats.norm.interval(alpha=0.95, 
                           loc=mean, 
                           scale=sigma / np.sqrt(n))
```
