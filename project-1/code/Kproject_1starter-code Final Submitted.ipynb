##### 12. Write a function using only list comprehensions, no loops, to compute standard deviation. Using this function, calculate the standard deviation of each numeric column in both data sets. Add these to a list called `sd`.

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

```python
# apply the List comprehension
sd=[np.std(sat_act[column]) for column in sat_act.columns[1:]]
sd
```

```python
def standard_deviation(lst):
    """ tocalculate the standard deviation of the list"""
    mean_value =np.mean(lst)
    inside_root = sum([(i-mean_value)**2 for i in lst])/len(lst)
    return np.sqrt(inside_root)
    
```

```python
sd = sat_act.iloc[:,1:].apply(standard_deviation)
sd
```

##### 27. Generate and interpret 95% confidence intervals for SAT and ACT participation rates.

```python
# Sat
mean=np.mean(SAT_part)
sigma=np.std(SAT_part, ddof=1)
n=len(SAT_part)

stats.norm.interval(alpha=0.95, 
                           loc=mean, 
                           scale=sigma / np.sqrt(n))
```

```python
# act
mean=np.mean(ACT_part)
sigma=np.std(ACT_part, ddof=1)
n=len(ACT_part)

stats.norm.interval(alpha=0.95, 
                           loc=mean, 
                           scale=sigma / np.sqrt(n))
```
